```python
def sum_nested_list(nested_list):
    total = 0
    for item in nested_list:
        if isinstance(item, list):
            # Recursively sum nested list
            total += sum_nested_list(item)
        else:
            total += item
    return total

# Test
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Sum of nested list:", result)
```

```python
def generate_permutations(s):
    if len(s) <= 1:
        return [s]

    permutations = set()  # use set to avoid duplicates
    for i, char in enumerate(s):
        # remaining string without the current char
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.add(char + perm)

    return list(permutations)

# Test
print(generate_permutations("abc"))
print(generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            # Recursively sum subdirectory
            total_size += calculate_directory_size(value)
        else:
            # File size in KB
            total_size += value
    return total_size

# Test
sample_directory = {
    "file1.txt": 10,
    "file2.txt": 20,
    "subdir1": {
        "file3.txt": 30,
        "subdir2": {
            "file4.txt": 40
        }
    },
    "subdir3": {
        "file5.txt": 50
    }
}

total = calculate_directory_size(sample_directory)
print("Total directory size (KB):", total)
```

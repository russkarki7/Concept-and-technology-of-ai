Task3

```python
# Missing values
hdi_2022_df.isnull().sum()
```

```python
# Replace special missing characters
hdi_2022_df.replace(["â€“", "-", ".."], np.nan, inplace=True)

# Convert numeric columns
numeric_cols = ['hdi', 'life expectancy', 'gross inc percap',
                'gender development', 'expected schooling', 'mean schooling']

for col in numeric_cols:
    if col in hdi_2022_df.columns:
        hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')

# Drop duplicates
hdi_2022_df.drop_duplicates(inplace=True)

# Handle missing values (dropping rows with missing HDI)
hdi_2022_df.dropna(subset=['hdi'], inplace=True)
```

Task4


```python
print("Mean HDI:", hdi_2022_df['hdi'].mean())
print("Median HDI:", hdi_2022_df['hdi'].median())
print("Std Dev HDI:", hdi_2022_df['hdi'].std())

# Highest & Lowest HDI
print("Highest HDI Country:")
hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax(), ['country', 'hdi']]

print("Lowest HDI Country:")
hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin(), ['country', 'hdi']]
```

Task5

```python
high_hdi = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]

high_hdi_sorted = high_hdi.sort_values(
    by='gross_inc_percap',
    ascending=False
)

high_hdi_sorted[['country', 'hdi', 'gross_inc_percap']].head(10)
```

Task6

```python
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi < 0.700:
        return "Medium"
    elif hdi < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(hdi_category)

hdi_2022_df['HDI Category'].value_counts()
```

```python
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

PROBLEM 2 â€“ Advanced HDI Exploration

Task 1: South Asia Subset

```python
south_asia = ["Afghanistan", "Bangladesh", "Bhutan", "India",
              "Maldives", "Nepal", "Pakistan", "Sri Lanka"]

hdi_sa = df[df['country'].isin(south_asia)].copy()
hdi_sa.to_csv("HDI_SouthAsia.csv", index=False)
```

Task 2: Composite Development Score

```python
hdi_sa['Composite Score'] = (
    0.30 * hdi_sa['life_expectancy'] +
    0.30 * hdi_sa['gross_inc_percap']
)

top5 = hdi_sa.sort_values(by='Composite Score', ascending=False).head(5)

plt.figure(figsize=(6,4))
plt.barh(top5['country'], top5['Composite Score'])
plt.title("Top 5 South Asian Countries by Composite Score")
plt.xlabel("Composite Score")
plt.show()
```

Task 3: Outlier Detection (IQR)

```python
Q1 = hdi_sa['hdi'].quantile(0.25)
Q3 = hdi_sa['hdi'].quantile(0.75)
IQR = Q3 - Q1

outliers = hdi_sa[
    (hdi_sa['hdi'] < Q1 - 1.5 * IQR) |
    (hdi_sa['hdi'] > Q3 + 1.5 * IQR)
]

plt.figure(figsize=(6,5))
sns.scatterplot(data=hdi_sa, x='gross_inc_percap', y='hdi')
sns.scatterplot(data=outliers, x='gross_inc_percap', y='hdi', color='red')
plt.title("Outliers in HDI vs GNI")
plt.xlabel("Gross Income per Capita")
plt.ylabel("HDI")
plt.show()
```

Task 4: Correlation Analysis

```python
# Make sure there are no extra spaces in column names
hdi_sa.columns = hdi_sa.columns.str.strip()

# Use the correct column names from your DataFrame
metrics = ['Gender Development', 'Life Expectancy']  # adjust these to match your exact column names

for m in metrics:
    if m in hdi_sa.columns:  # check if column exists
        corr = hdi_sa[m].corr(hdi_sa['hdi'])
        print(f"Correlation between HDI and {m}: {corr:.3f}")

        sns.regplot(data=hdi_sa, x=m, y='hdi')
        plt.title(f"HDI vs {m}")
        plt.show()
    else:
        print(f"Column '{m}' not found in the DataFrame.")
```

Task 5: GNIâ€“HDI Gap

```python
import matplotlib.pyplot as plt

# Correct column names based on your DataFrame
gross_income_col = 'gross_inc_percap'
hdi_col = 'hdi'
country_col = 'country'

# Create the GNIâ€“HDI gap column
hdi_sa['GNI_HDI_Gap'] = hdi_sa[gross_income_col] - hdi_sa[hdi_col]

# Top positive gaps
top_pos = hdi_sa.sort_values(by='GNI_HDI_Gap', ascending=False).head(3)
# Top negative gaps
top_neg = hdi_sa.sort_values(by='GNI_HDI_Gap').head(3)

# Plot top positive gaps
plt.figure(figsize=(6,4))
plt.bar(top_pos[country_col], top_pos['GNI_HDI_Gap'], color='green')
plt.title("Top Positive GNIâ€“HDI Gaps")
plt.ylabel("GNIâ€“HDI Gap")
plt.show()

# Plot top negative gaps
plt.figure(figsize=(6,4))
plt.bar(top_neg[country_col], top_neg['GNI_HDI_Gap'], color='red')
plt.title("Top Negative GNIâ€“HDI Gaps")
plt.ylabel("GNIâ€“HDI Gap")
plt.show()
```

PROBLEM 3 â€“ South Asia vs Middle East

```python
middle_east = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan",
               "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
               "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]

sa_20_22 = hdi_2020_22[hdi_2020_22['country'].isin(south_asia)]
me_20_22 = hdi_2020_22[hdi_2020_22['country'].isin(middle_east)]

sa_20_22.to_csv("HDI_SouthAsia_2020_2022.csv", index=False)
me_20_22.to_csv("HDI_MiddleEast_2020_2022.csv", index=False)
```

Descriptive Statistics

```python
print("South Asia Mean HDI:", sa_20_22['hdi'].mean())
print("Middle East Mean HDI:", me_20_22['hdi'].mean())
```

HDI Variation

```python
for name, data in zip(["South Asia", "Middle East"], [sa_20_22, me_20_22]):
    hdi_range = data['hdi'].max() - data['hdi'].min()
    cv = data['hdi'].std() / data['hdi'].mean()
    print(f"{name} â†’ Range: {hdi_range:.3f}, CV: {cv:.3f}")
```
